```python
import pandas as pd
import numpy as np
```

```python
html_list=pd.read_html('https://worldpopulationreview.com/country-rankings/freedom-index-by-country')
len(html_list)
```

```python
table=html_list[2]
table
```

```python
html_list=pd.read_html("https://worldpopulationreview.com/country-rankings/most-charitable-countries")
len(html_list)
```

```python
table2=html_list[0]
table2
```

```python
mergedtables = pd.merge(table, table2, how='left', on='Country')
```

```python
mergedtables
```

```python
html_list3=pd.read_html('https://en.wikipedia.org/wiki/Satisfaction_with_Life_Index')
len(html_list)
```

```python
table3=html_list3[1]
table3
```

```python
# Concatenate the data from the 'Rank', 'Country', and 'SWL' columns
new_df = pd.concat([table3[['Rank', 'Country', 'SWL']], table3[['Rank.1', 'Country.1', 'SWL.1']].rename(columns={'Rank.1': 'Rank', 'Country.1': 'Country', 'SWL.1': 'SWL'})], ignore_index=True)

# Display the reformatted DataFrame
print(new_df)
```

```python
new_df
```

```python
merged_data= pd.merge(mergedtables, new_df, how='left', on='Country')
```

```python
merged_data
```

```python
file_path = "all_world_data.csv"
merged_data.to_csv(file_path, index=False)
```
