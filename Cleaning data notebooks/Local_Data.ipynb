```python
import pandas as pd
import numpy as np
```

```python
wiki_list=pd.read_html('https://en.wikipedia.org/wiki/List_of_London_boroughs')
len(wiki_list)
```

```python
table=wiki_list[0]
table=table[['Borough','Political control','Area (sq mi)','Population (2019 est)','Co-ordinates']]
table.columns = ["Borough", "Political Control","Area","Population","Co-ordinates"]
rename_dict = {
    'Barking and Dagenham[note 1]': 'Barking and Dagenham',
    'Greenwich[note 2]': 'Greenwich',
    'Hammersmith and Fulham[note 4]': 'Hammersmith and Fulham'
}
table['Borough'].replace(rename_dict, inplace=True)
table
```

```python
crime_data = pd.read_csv('/content/MPS Borough Level Crime (most recent 24 months).csv')
numerical_columns = crime_data.columns[3:]
crime_data['Total'] = crime_data[numerical_columns].sum(axis=1)
crime_data.drop(columns=numerical_columns, inplace=True)
column = 'MinorText'
crime_data.drop(columns=column, inplace=True)
crime_data.columns=["Type","Borough","Total"]
crime_data = crime_data.pivot_table(index='Borough', columns='Type', values='Total', aggfunc='sum')
crime_data.reset_index(inplace=True)
numerical_columns = crime_data.columns[1:]
crime_data['Total Crime'] = crime_data[numerical_columns].sum(axis=1)
crime_data
```

```python
datamerged = pd.merge(table, crime_data, how='left', on='Borough')
```

```python
datamerged
```

```python
local_authority_data = pd.read_csv('/content/Local_Authority.csv')
new_column_name = 'Borough'
local_authority_data = local_authority_data.rename(columns={'LAD22NM': new_column_name})
```

```python
datamerged = datamerged = pd.merge(datamerged, local_authority_data, how='left', on='Borough')
datamerged
```

```python
happiness_data = pd.read_csv('/content/london_happiness2.csv')
column_name = 'Unnamed: 2'
happiness_data = happiness_data[happiness_data[column_name] == 'Happiness']
columns_to_drop = ['Personal well-being estimates by local authority', 'Unnamed: 2', 'Unnamed: 3','Unnamed: 5','Unnamed: 6','Unnamed: 7']
happiness_data = happiness_data.drop(columns=columns_to_drop)
happiness_data.columns=['Borough','Average Happiness Score']
happiness_data
```

```python
datamerged = datamerged = pd.merge(datamerged, happiness_data, how='left', on='Borough')
```
